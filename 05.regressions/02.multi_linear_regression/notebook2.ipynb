```python
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import f_regression
```

```python
data = pd.read_csv("multi_linear_regression.csv")
```

```python
x = data[["SAT", "Rand 1,2,3"]]
y = data["GPA"]
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
r2 = reg.score(x,y)
n = x.shape[0]
p = x.shape[1]
adj_r2 = 1-(1-r2)*(n-1)/(n-p-1)
```

```python
print("Intercept:", reg.intercept_)
print("Coefficient:", reg.coef_)
print("R2:", r2)
print("Adjusted R2:", adj_r2)
```

1. e⁻¹¹ = 1/10¹¹
2. 0.000 represents the p value of SAT
3. 0.676 represents the p value of Rand 1,2,3
4. Why p value (or field) of Rand 1,2,3 is not useful

```python
f_reg = f_regression(x,y)
f_stastistics = f_reg[0]
p_values = f_reg[1].round(3)
print("F Statistics:", f_stastistics)
print("P Values:", p_values)
```

```python
reg_summary = pd.DataFrame(data=x.columns.values, columns=["Features"])
reg_summary["Coefficients"] = reg.coef_
reg_summary["p-values"] = p_values
print(reg_summary)
```
